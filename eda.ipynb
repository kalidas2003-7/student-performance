```python
import pandas as pd
df=pd.read_csv("StudentsPerformance.csv")
df
```

```python
overall_avg= df[['math score', 'reading score', 'writing score']].values.mean()
overall_avg
```

```python
df['overall_avg_score']= (df['math score'] + df['reading score'] + df['writing score'])/3
```

```python
df
```

```python
df.drop(columns=['math score', 'reading score', 'writing score'], inplace=True)
```

```python
df
```

```python
df.drop(columns=['overall score'], inplace=True)
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
gender_score = df.groupby("gender")["overall_avg_score"].mean()
```

```python
plt.figure()
plt.bar(gender_score.index, gender_score.values)
plt.xlabel("Gender",fontsize=12)
plt.ylabel("overall_avg_ Score")
plt.title("Average Score by Gender")
plt.show()
```

```python
male_scores = df[df["gender"] == "male"]["overall_avg_score"]
female_scores = df[df["gender"] == "female"]["overall_avg_score"]
```

```python
plt.figure()
plt.hist(male_scores, bins=10, alpha=0.7, label="Male")
plt.hist(female_scores, bins=10, alpha=0.7, label="Female")
plt.xlabel("Average Score")
plt.ylabel("Frequency")
plt.title("Histogram of Overall Average Score by Gender")
plt.legend()
plt.show()
```

```python
import seaborn as sns
plt.figure()
sns.histplot(
    data=df,
    x="overall_avg_score",
    hue="gender",
    bins=20,
    kde=True,
    multiple="layer"
)

plt.xlabel("Average Score")
plt.ylabel("Count")
plt.title("Distribution of Average Score by Gender")
plt.show()
```

```python
sns.displot(df, x="overall_avg_score", col="race/ethnicity", bins=15, kde=True)
```

```python
race_avg = df.groupby("race/ethnicity")["overall_avg_score"].mean()
 
```

```python
plt.figure()
plt.bar(race_avg.index, race_avg.values)
plt.xlabel("Race/Ethnicity")
plt.ylabel("Overall Average Score")
plt.title("Average Score by Race/Ethnicity")
plt.show()
```

```python
plt.figure()
sns.histplot(
    data=df,
    x="overall_avg_score",
    hue="race/ethnicity",
    bins=20,
    kde=True,
    multiple="layer"
)

plt.xlabel("Average Score")
plt.ylabel("race/ethnicity")
plt.title("Distribution of Average Score by Race/Ethnicity")
plt.show()
```
